```python
import pandas as pd
import numpy as np
dtype_dict = {'bathrooms':float, 'waterfront':int, 'sqft_above':int, 'sqft_living15':float, 'grade':int, 'yr_renovated':int, 'price':float, 'bedrooms':float, 'zipcode':str, 'long':float, 'sqft_lot15':float, 'sqft_living':float, 'floors':float, 'condition':int, 'lat':float, 'date':str, 'sqft_basement':int, 'yr_built':int, 'id':str, 'sqft_lot':int, 'view':int}

sales = pd.read_csv('kc_house_data.csv', dtype=dtype_dict)
sales = sales.sort_values(['sqft_living','price'])
```

```python
l2_small_penalty = 1.5e-5
```

```python
from sklearn import linear_model
import numpy as np

def polynomial_frame(feature,degree):
    poly_data=pd.DataFrame()
    for i in range(1,degree+1):
        name='pwr'+str(i)
        poly_data[name]=feature**i
    return poly_data

poly15_data = polynomial_frame(sales['sqft_living'], 15) # use equivalent of `polynomial_sframe`
model = linear_model.Ridge(alpha=l2_small_penalty, normalize=True)
model.fit(poly15_data, sales['price'])
model.coef_
```

```python
train_valid_shuffled = pd.read_csv('wk3_kc_house_train_valid_shuffled.csv', dtype=dtype_dict)
test = pd.read_csv('wk3_kc_house_test_data.csv', dtype=dtype_dict)
```

```python
n = len(train_valid_shuffled)
k = 10 # 10-fold cross-validation

for i in range(k):
    start = int((n*i)/k)
    end = int((n*(i+1))/k-1)
    print (i, (start, end))
```

```python
def k_fold_cross_validation(k, l2_penalty, data):
    n=len(train_valid_shuffled)
    rsslist=[]
    for i in range(k):
        start=int((n*i)/k)
        end=int((n*(i+1))/k-1)
        valid=data[start:end+1]
        train=data[end+1:n].append(data[0:start])
        train_sq=polynomial_frame(train['sqft_living'],15)
        valid_sq=polynomial_frame(valid['sqft_living'],15)
        feat=train_sq.columns
        val_feat=valid_sq.columns
        train_sq['price']=train['price']
        valid_sq['price']=valid['price']
        kfold_model=linear_model.Ridge(alpha=l2_penalty,normalize=True)
        kfold_model.fit(train_sq[feat],train[['price']])
        rss=(((kfold_model.predict(valid_sq[val_feat]))-valid_sq[['price']])**2).sum()
        rsslist.append(rss)
    mean_rss=np.mean(rsslist)
    return mean_rss
```

```python
p_list=[]
for i in range(6,19):
    p_list.append(int(10**(i/2)))
p_list
```

```python
lambda_list=[]
for i in p_list:
    lambda_list.append(k_fold_cross_validation(10,i,train_valid_shuffled))
lambda_list.index(min(lambda_list))
```
